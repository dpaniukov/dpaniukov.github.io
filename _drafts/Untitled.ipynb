```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def f(x):
    arr=np.random.uniform(0,1,size=x)
    plt.plot(range(x), arr)
    plt.show()
    return arr
interact(f, x=(1,20,1));
```
